```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import math
import scipy.stats as stats
from scipy.stats import fisher_exact
```

```python
df = pd.read_csv('1002033072.csv')
```

```python
df
```

```python
df.info()
```

```python
df1 = pd.DataFrame(columns=['SNP', 'P_Value', 'CorrectedP_Value', 'Significant PValue', 'Significant CorrectedPValue'])
        
oddslist = []
reslist  = []
benclist = []

# TODO: add the corrected P value to the results as well
for i in range(1000):
    a = df.Case_Num_T_Allele.iloc[[i]].values[0]
    x = df.Case_Num_C_Allele.iloc[[i]].values[0]
    b = df.Control_Num_T_Allele.iloc[[i]].values[0]
    y = df.Control_Num_C_Allele.iloc[[i]].values[0]
    
    contingencymatrix = [[a,x], [b,y]]
    odds, res = fisher_exact(contingencymatrix)
    if(res < 5e-8):
        res_significant = True
    else:
        res_significant = False 
    
    oddslist.append(odds)
    reslist.append(res)
    
    #Corrected P value
    benc = res / 1000
    benclist.append(benc)
    
    if(benc < 5e-8) :
        bonferroni_significant = True
    else:
        bonferroni_significant = False 

    df1 = df1.append({
        'SNP': 'snp_' + str(i),
        'P_Value': res,
        'CorrectedP_Value': benc,
        'Significant PValue': res_significant,
        'Significant CorrectedPValue': bonferroni_significant}, ignore_index=True)
```

```python
# The fishers test basically has a pvalue that tells us the probability that we encountered
# the following contingency table randomly with the probability in the 'pvalue'
# Therefore, to facilitate that, we use the two sided alternative for the test
df1
```

```python
df1.to_csv('results.csv', index=False)
```
